```python
# 연령대 별 총시청률
age = TV_daily_0.groupby("Target_Audience").size().reset_index(name='count')
age.sort_values('count', ascending=False, inplace=True)

age = TV_daily_0.groupby('Target_Audience').agg('sum')[['GRP']].round().reset_index()
age.sort_values('GRP', ascending=False, inplace=True)

# 연령대 중 "개인" 제거
age = age.drop([0])
age.reset_index(drop=True, inplace=True)

age
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(data=age, x='Target_Audience', y='GRP', palette='Blues_d')
plt.title('연령대별 총 시청률', fontsize=15)

plt.show()
```
